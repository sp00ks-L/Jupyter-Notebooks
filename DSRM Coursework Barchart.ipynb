```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
plt.style.use('ggplot')
xs = [x for x in range(0, 6)]
# Dummy data
ys = [-400000, 200000, 500000, 300000, -20700000, 2500000] 
plt.ylim(-25000000, 5000000)
plt.title("Job Losses and gains since Dec-2019\nTotal nonfarm payrolls, change from previous month\n(Not real data)")

# If val positive == blue else red
f = lambda x: '#e24a33' if x < 0 else '#348abd'
colours = [f(x) for x in ys]

sns.barplot(x=xs, y=ys, palette=colours)
plt.xticks(xs, labels=["Dec-2019", "Jan-2020", "Feb-2020", "Mar-2020", "Apr-2020", "May-2020"], rotation=45)
plt.yticks([val * 1e6 for val in range(-25, 11, 5)], labels=["-25", "-20", "-15", "-10", "-5", "0", "5"])
plt.ylabel("Change in Millions", rotation=0, labelpad = 60)
plt.xlabel("Month-Year", labelpad=20)




plt.show()
```
